```python
try:

    client_id = dbutils.secrets.get(
        scope="azure-kv-scope",
        key="client-id"
    )

    tenant_id = dbutils.secrets.get(
        scope="azure-kv-scope",
        key="tenant-id"
    )

    client_secret = dbutils.secrets.get(
        scope="azure-kv-scope",
        key="client-secret"
    )

    storage_account = dbutils.secrets.get(
        scope="azure-kv-scope",
        key="storage-account"
    )

    spark.conf.set(
        f"fs.azure.account.auth.type.{storage_account}.dfs.core.windows.net",
        "OAuth"
    )

    spark.conf.set(
        f"fs.azure.account.oauth.provider.type.{storage_account}.dfs.core.windows.net",
        "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider"
    )

    spark.conf.set(
        f"fs.azure.account.oauth2.client.id.{storage_account}.dfs.core.windows.net",
        client_id
    )

    spark.conf.set(
        f"fs.azure.account.oauth2.client.secret.{storage_account}.dfs.core.windows.net",
        client_secret
    )

    spark.conf.set(
        f"fs.azure.account.oauth2.client.endpoint.{storage_account}.dfs.core.windows.net",
        f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
    )

    # Validation

    test_path = (
        f"abfss://dev@{storage_account}.dfs.core.windows.net/"
    )

    dbutils.fs.ls(test_path)

    print("================================")
    print("BIB ADLS Connection Successful")
    print("Storage Account :", storage_account)
    print("================================")

except Exception as e:

    print("================================")
    print("BIB ADLS Connection Failed")
    print(str(e))
    print("================================")

    raise e
```
